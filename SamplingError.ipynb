```python
import pandas as pd
```

```python
dataframe=pd.read_csv('airlines_data_part.CSV')
```

```python
dataframe11=dataframe.sample(20000)
```

```python
dataframe11
```

```python
import matplotlib.pyplot as plt
sales=dataframe11['sales']
months=dataframe11['month']
plt.bar(months,sales)
plt.ylabel('Sales')
plt.xlabel('Months')
plt.show()
```
